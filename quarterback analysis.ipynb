```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

all_players_df = pd.read_csv('final_df.csv')
del all_players_df['Unnamed: 0']

all_players_df.head()
```

```python
quarterback_df = all_players_df[all_players_df['Pos_x'] == 'QB']

quarterback_df.info()
```

```python
corr = np.corrcoef(np.random.randn(10, 200))
mask = np.zeros_like(corr)
mask[np.triu_indices_from(mask)] = True
draft_with_pb = quarterback_df[['Round','Wt', '40yd', 'Vertical', 'Broad Jump', '3Cone', 'Shuttle','Pick', 'Year','PB']]
                   

plt.figure(figsize=(10,7))
sns.heatmap(draft_with_pb.corr(), annot=True,mask=mask,cmap="Blues")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB


y = quarterback_df['PB'] > 0
features = ['C Games','C Pass Cmp', 'C Pass Att', 'C Pass Pct', 'C Pass Yds', 'C Pass Y/A', 'C Pass AY/A', 'C Pass Int','Round','Wt', '40yd', 'Vertical', 'Broad Jump', '3Cone', 'Shuttle','Pick']
X = quarterback_df[features].fillna(0)

train_X, test_X, train_y, test_y = train_test_split(X, y,test_size=.4, random_state=22)
te_model = GaussianNB()

te_model.fit(train_X, train_y)
confidence = te_model.score(test_X,test_y)
print('The confidence of the model is :',confidence)

y_pred = te_model.predict(test_X)
print("Number of mislabeled points out of a total %d points : %d"
      % (test_X.shape[0], (test_y != y_pred).sum()))
```
