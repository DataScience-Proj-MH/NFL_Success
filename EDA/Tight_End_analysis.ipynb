```python
# Code you have previously used to load data
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn import svm , neighbors


#te_df = te_df.loc[te_df['Year'] < 2012]
te_df = all_players_df[all_players_df['Pos_x'] == 'WR']

y =te_df['Round']
features = ['C Games','C Rec', 'C RecYds', 'C  RecAvg', 'C RecTD', 'C ScrimPLays', 'C ScrimYds','C ScrimAvg', 'C ScrimTD','Wt', '40yd', 'Vertical', 'Broad Jump', '3Cone', 'Shuttle','Pick', 'Year']
X = te_df[features].fillna(0)

# Split into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y,test_size=.3, random_state=4323334)
te_model = neighbors.KNeighborsClassifier()

# fit your model
te_model.fit(train_X, train_y)
#print confidence
confidence = te_model.score(val_X , val_y)
print(confidence)
y_predict = te_model.predict(val_X)
print(y_predict)
```
